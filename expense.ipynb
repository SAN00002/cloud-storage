```python
import pandas as pd

df=pd.read_excel("/expenses.xlsx")
df.head()
df.isnull().sum()
df.dropna(inplace=True)
df.dtypes
df['Date'] = pd.to_datetime(df['Date'])
df.drop_duplicates(inplace=True)
df['Category'] = df['Category'].str.strip().str.title()
df['Week'] = df['Date'].dt.isocalendar().week
df['Month'] = df['Date'].dt.month
df = df.sort_values(by='Date')
df.to_csv("cleaned_expenses.csv", index=False)
print(df.describe(include='all')
)
print(df['Category'].value_counts())
df['Category'].value_counts().plot(kind='barh', title='Spending by Category')
weekly = df.groupby('Week')['Amount (â‚¹)'].sum()
weekly.plot(kind='line', marker='o', title='Weekly Expense Trend')

import matplotlib.pyplot as plt
plt.xlabel('Week Number')
plt.ylabel('Total Spending (â‚¹)')
plt.grid(True)
plt.tight_layout()
plt.show()
df['Category'].value_counts().plot(kind='bar', title='Spending by Category')

plt.xlabel("Category")
plt.ylabel("Number of Entries")
plt.tight_layout()
plt.show()

df[df['Recurring?'] == 'Yes']['Category'].value_counts()
df[df['Recurring?'] == 'Yes']['Category'].value_counts().plot(kind='bar')

```
